```python
# ==========================================
# IMPORT
# ==========================================

from dataclasses import dataclass, field
from typing import List


# ==========================================
# CLASS BUKU
# ==========================================

@dataclass
class Buku:

    # Attribute / data buku
    kode_buku: str
    judul: str
    penulis: str
    stok: int

    # Method untuk menampilkan info buku
    def tampil_info(self):
        print("=== DATA BUKU ===")
        print(f"Kode Buku : {self.kode_buku}")
        print(f"Judul     : {self.judul}")
        print(f"Penulis   : {self.penulis}")
        print(f"Stok      : {self.stok}")


# ==========================================
# CLASS ANGGOTA
# ==========================================

@dataclass
class Anggota:

    # Attribute anggota
    id_anggota: str
    nama: str
    alamat: str

    # Method menampilkan info anggota
    def tampil_info(self):
        print("=== DATA ANGGOTA ===")
        print(f"ID Anggota : {self.id_anggota}")
        print(f"Nama       : {self.nama}")
        print(f"Alamat     : {self.alamat}")


# ==========================================
# CLASS DETAIL PEMINJAMAN
# ==========================================

@dataclass
class DetailPeminjaman:

    # Association ke class Buku
    buku: Buku

    # Jumlah buku dipinjam
    jumlah: int

    # Method tampil detail
    def tampil_detail(self):
        print("=== DETAIL PEMINJAMAN ===")
        print(f"Buku   : {self.buku.judul}")
        print(f"Jumlah : {self.jumlah}")


# ==========================================
# CLASS TRANSAKSI PEMINJAMAN
# ==========================================

@dataclass
class TransaksiPeminjaman:

    # Attribute transaksi
    kode_transaksi: str

    # Association ke Anggota
    anggota: Anggota

    # Composition ke DetailPeminjaman
    detail_pinjam: List[DetailPeminjaman] = field(default_factory=list)

    # Menambah detail peminjaman
    def tambah_detail(self, detail: DetailPeminjaman):
        self.detail_pinjam.append(detail)

    # Menampilkan transaksi
    def tampil_transaksi(self):

        print("\n=== TRANSAKSI PEMINJAMAN ===")
        print(f"Kode Transaksi : {self.kode_transaksi}")
        print(f"Nama Anggota   : {self.anggota.nama}")

        print("\nDaftar Buku Dipinjam:")

        # Loop detail peminjaman
        for detail in self.detail_pinjam:
            print(f"- {detail.buku.judul} ({detail.jumlah} buku)")


# ==========================================
# CLASS PERPUSTAKAAN
# ==========================================

class Perpustakaan:

    # Constructor
    def __init__(self):

        # Aggregation daftar buku
        self.daftar_buku = []

        # Aggregation daftar anggota
        self.daftar_anggota = []

        # Menyimpan transaksi
        self.daftar_transaksi = []

    # Method tambah buku
    def tambah_buku(self, buku: Buku):
        self.daftar_buku.append(buku)

    # Method tambah anggota
    def tambah_anggota(self, anggota: Anggota):
        self.daftar_anggota.append(anggota)

    # Method tambah transaksi
    def tambah_transaksi(self, transaksi: TransaksiPeminjaman):
        self.daftar_transaksi.append(transaksi)

    # Menampilkan semua data
    def tampil_data(self):

        print("\n===== DATA BUKU =====")

        for buku in self.daftar_buku:
            print(f"{buku.kode_buku} - {buku.judul}")

        print("\n===== DATA ANGGOTA =====")

        for anggota in self.daftar_anggota:
            print(f"{anggota.id_anggota} - {anggota.nama}")

        print("\n===== DATA TRANSAKSI =====")

        for transaksi in self.daftar_transaksi:
            transaksi.tampil_transaksi()


# ==========================================
# MEMBUAT OBJECT
# ==========================================

# Object buku
buku1 = Buku(
    "B001",
    "Python Dasar",
    "Andi",
    10
)

buku2 = Buku(
    "B002",
    "OOP Python",
    "Budi",
    5
)

# Object anggota
anggota1 = Anggota(
    "A001",
    "Rizky",
    "Sidoarjo"
)

# Object detail peminjaman
detail1 = DetailPeminjaman(
    buku1,
    2
)

detail2 = DetailPeminjaman(
    buku2,
    1
)

# Object transaksi
transaksi1 = TransaksiPeminjaman(
    "T001",
    anggota1
)

# Menambahkan detail ke transaksi
transaksi1.tambah_detail(detail1)
transaksi1.tambah_detail(detail2)

# Object perpustakaan
perpus = Perpustakaan()

# Menambahkan data
perpus.tambah_buku(buku1)
perpus.tambah_buku(buku2)

perpus.tambah_anggota(anggota1)

perpus.tambah_transaksi(transaksi1)

# ==========================================
# OUTPUT
# ==========================================

perpus.tampil_data()
```

HUBUNGAN ANTAR CLASS

Perpustakaan > buku, anggota, transaksi, = Aggregation

Transaksi > anggota = association

Transaksi > detail peminjaman = composition

Detail peminjaman > buku = association

```python

# ==========================================
# IMPORT LIBRARY
# ==========================================

# dataclass digunakan untuk membuat class lebih singkat
from dataclasses import dataclass, field

# typing digunakan untuk type hints / generics
from typing import List, Dict


# ==========================================
# CLASS BUKU
# ==========================================

@dataclass
class Buku:

    # Attribute buku
    kode_buku: str
    judul: str
    penulis: str
    stok: int

    # Method menampilkan info buku
    def tampil_info(self):

        print("\n=== DATA BUKU ===")
        print(f"Kode Buku : {self.kode_buku}")
        print(f"Judul     : {self.judul}")
        print(f"Penulis   : {self.penulis}")
        print(f"Stok      : {self.stok}")


# ==========================================
# CLASS ANGGOTA
# ==========================================

@dataclass
class Anggota:

    # Attribute anggota
    id_anggota: str
    nama: str
    alamat: str

    # Method tampil info anggota
    def tampil_info(self):

        print("\n=== DATA ANGGOTA ===")
        print(f"ID Anggota : {self.id_anggota}")
        print(f"Nama       : {self.nama}")
        print(f"Alamat     : {self.alamat}")


# ==========================================
# CLASS DETAIL PEMINJAMAN
# ==========================================

@dataclass
class DetailPeminjaman:

    # Association ke class Buku
    buku: Buku

    # Jumlah buku dipinjam
    jumlah: int

    # Method tampil detail
    def tampil_detail(self):

        print("\n=== DETAIL PEMINJAMAN ===")
        print(f"Judul Buku : {self.buku.judul}")
        print(f"Jumlah     : {self.jumlah}")


# ==========================================
# CLASS TRANSAKSI PEMINJAMAN
# ==========================================

@dataclass
class TransaksiPeminjaman:

    # Attribute transaksi
    kode_transaksi: str

    # Association ke class Anggota
    anggota: Anggota

    # Composition ke DetailPeminjaman
    detail_pinjam: List[DetailPeminjaman] = field(default_factory=list)

    # Method tambah detail peminjaman
    def tambah_detail(self, detail: DetailPeminjaman):

        self.detail_pinjam.append(detail)

    # Method tampil transaksi
    def tampil_transaksi(self):

        print("\n=== DATA TRANSAKSI ===")
        print(f"Kode Transaksi : {self.kode_transaksi}")
        print(f"Nama Anggota   : {self.anggota.nama}")

        print("\nDaftar Buku Dipinjam:")

        # Menampilkan semua detail
        for detail in self.detail_pinjam:

            print(
                f"- {detail.buku.judul} "
                f"({detail.jumlah} buku)"
            )


# ==========================================
# CLASS PERPUSTAKAAN
# ==========================================

class Perpustakaan:

    # Constructor
    def __init__(self):

        # List khusus object Buku
        self.daftar_buku: List[Buku] = []

        # Dictionary anggota
        # key = id anggota
        # value = object anggota
        self.daftar_anggota: Dict[str, Anggota] = {}

        # List transaksi
        self.daftar_transaksi: List[TransaksiPeminjaman] = []

    # ======================================
    # METHOD TAMBAH BUKU
    # ======================================

    def tambah_buku(self, buku: Buku):

        self.daftar_buku.append(buku)

        print(f"\nBuku '{buku.judul}' berhasil ditambahkan")

    # ======================================
    # METHOD TAMBAH ANGGOTA
    # ======================================

    def tambah_anggota(self, anggota: Anggota):

        self.daftar_anggota[anggota.id_anggota] = anggota

        print(f"\nAnggota '{anggota.nama}' berhasil ditambahkan")

    # ======================================
    # METHOD TAMBAH TRANSAKSI
    # ======================================

    def tambah_transaksi(
        self,
        transaksi: TransaksiPeminjaman
    ):

        self.daftar_transaksi.append(transaksi)

        print(
            f"\nTransaksi "
            f"'{transaksi.kode_transaksi}' "
            f"berhasil ditambahkan"
        )

    # ======================================
    # METHOD TAMPIL BUKU
    # ======================================

    def tampil_buku(self):

        print("\n===== DAFTAR BUKU =====")

        for buku in self.daftar_buku:

            print(
                f"{buku.kode_buku} - "
                f"{buku.judul}"
            )

    # ======================================
    # METHOD TAMPIL ANGGOTA
    # ======================================

    def tampil_anggota(self):

        print("\n===== DAFTAR ANGGOTA =====")

        for anggota in self.daftar_anggota.values():

            print(
                f"{anggota.id_anggota} - "
                f"{anggota.nama}"
            )

    # ======================================
    # METHOD TAMPIL TRANSAKSI
    # ======================================

    def tampil_transaksi(self):

        print("\n===== DAFTAR TRANSAKSI =====")

        for transaksi in self.daftar_transaksi:

            transaksi.tampil_transaksi()


# ==========================================
# MEMBUAT OBJECT
# ==========================================

# Membuat object perpustakaan
perpus = Perpustakaan()


# ==========================================
# MEMBUAT OBJECT BUKU
# ==========================================

buku1 = Buku(
    "B001",
    "Python Dasar",
    "Andi",
    10
)

buku2 = Buku(
    "B002",
    "Pemrograman OOP",
    "Budi",
    5
)


# ==========================================
# MEMBUAT OBJECT ANGGOTA
# ==========================================

anggota1 = Anggota(
    "A001",
    "Rizky",
    "Sidoarjo"
)


# ==========================================
# MENAMBAHKAN DATA KE PERPUSTAKAAN
# ==========================================

perpus.tambah_buku(buku1)
perpus.tambah_buku(buku2)

perpus.tambah_anggota(anggota1)


# ==========================================
# MEMBUAT DETAIL PEMINJAMAN
# ==========================================

detail1 = DetailPeminjaman(
    buku1,
    2
)

detail2 = DetailPeminjaman(
    buku2,
    1
)


# ==========================================
# MEMBUAT TRANSAKSI
# ==========================================

transaksi1 = TransaksiPeminjaman(
    "T001",
    anggota1
)


# ==========================================
# MENAMBAHKAN DETAIL KE TRANSAKSI
# ==========================================

transaksi1.tambah_detail(detail1)
transaksi1.tambah_detail(detail2)


# ==========================================
# MENYIMPAN TRANSAKSI
# ==========================================

perpus.tambah_transaksi(transaksi1)


# ==========================================
# MENAMPILKAN SEMUA DATA
# ==========================================

perpus.tampil_buku()

perpus.tampil_anggota()

perpus.tampil_transaksi()
```

Alur Program
1. Input Data Buku

2. Input Data Anggota

Pengguna memasukkan:
- ID anggota
- nama anggota
- alamat

Lalu sistem:
- membuat object Anggota
- menyimpan ke daftar anggota

3. Transaksi Peminjaman

Pengguna memasukkan:
- kode transaksi
- kode buku yang dipinjam
- jumlah pinjam

Lalu sistem:
- mencari buku berdasarkan kode
- membuat Detail Peminjaman
- menyimpan ke Transaksi Peminjaman

```python
# ==========================================
# IMPORT LIBRARY
# ==========================================

from dataclasses import dataclass, field
from typing import List


# ==========================================
# CLASS BUKU
# ==========================================

@dataclass
class Buku:

    kode_buku: str
    judul: str
    penulis: str
    stok: int


# ==========================================
# CLASS ANGGOTA
# ==========================================

@dataclass
class Anggota:

    id_anggota: str
    nama: str
    alamat: str


# ==========================================
# CLASS DETAIL PEMINJAMAN
# ==========================================

@dataclass
class DetailPeminjaman:

    buku: Buku
    jumlah: int


# ==========================================
# CLASS TRANSAKSI PEMINJAMAN
# ==========================================

@dataclass
class TransaksiPeminjaman:

    kode_transaksi: str
    anggota: Anggota

    detail_pinjam: List[DetailPeminjaman] = field(default_factory=list)

    # Method tambah detail
    def tambah_detail(self, detail):

        self.detail_pinjam.append(detail)

    # Method tampil transaksi
    def tampil_transaksi(self):

        print("\n===== DATA TRANSAKSI =====")
        print(f"Kode Transaksi : {self.kode_transaksi}")
        print(f"Nama Anggota   : {self.anggota.nama}")

        print("\nDaftar Buku Dipinjam:")

        for detail in self.detail_pinjam:

            print(
                f"- {detail.buku.judul} "
                f"({detail.jumlah} buku)"
            )


# ==========================================
# CLASS PERPUSTAKAAN
# ==========================================

class Perpustakaan:

    def __init__(self):

        self.daftar_buku: List[Buku] = []
        self.daftar_anggota: List[Anggota] = []
        self.daftar_transaksi: List[TransaksiPeminjaman] = []

    # Tambah buku
    def tambah_buku(self, buku):

        self.daftar_buku.append(buku)

    # Tambah anggota
    def tambah_anggota(self, anggota):

        self.daftar_anggota.append(anggota)

    # Tambah transaksi
    def tambah_transaksi(self, transaksi):

        self.daftar_transaksi.append(transaksi)

    # Tampilkan buku
    def tampil_buku(self):

        print("\n===== DAFTAR BUKU =====")

        for buku in self.daftar_buku:

            print(
                f"{buku.kode_buku} - "
                f"{buku.judul} - "
                f"Stok: {buku.stok}"
            )

    # Tampilkan anggota
    def tampil_anggota(self):

        print("\n===== DATA ANGGOTA =====")

        for anggota in self.daftar_anggota:

            print(
                f"{anggota.id_anggota} - "
                f"{anggota.nama}"
            )


# ==========================================
# PROGRAM UTAMA
# ==========================================

print("===================================")
print("    HALO PETUGAS PERPUSTAKAAN")
print(" SELAMAT DATANG DI UNUSIDALAB")
print("===================================")

# Membuat object perpustakaan
perpus = Perpustakaan()


# ==========================================
# INPUT DATA BUKU
# ==========================================

print("\n===== INPUT DATA BUKU =====")

# Validasi jumlah buku
while True:

    try:

        jumlah_buku = int(
            input("Berapa buku yang ingin ditambahkan? ")
        )

        break

    except ValueError:

        print("Harap isi dengan angka!")


# Loop input buku
for i in range(jumlah_buku):

    print(f"\nInput Buku ke-{i+1}")

    kode = input("Kode Buku : ")
    judul = input("Judul Buku : ")
    penulis = input("Penulis : ")

    # Validasi stok
    while True:

        try:

            stok = int(input("Stok Buku : "))

            break

        except ValueError:

            print("Harap isi stok dengan angka!")

    # Membuat object buku
    buku = Buku(
        kode,
        judul,
        penulis,
        stok
    )

    # Simpan buku
    perpus.tambah_buku(buku)

print("\nData buku berhasil ditambahkan!")


# ==========================================
# INPUT DATA ANGGOTA
# ==========================================

print("\n===== INPUT DATA ANGGOTA =====")

id_anggota = input("ID Anggota : ")
nama = input("Nama Anggota : ")
alamat = input("Alamat : ")

# Membuat object anggota
anggota = Anggota(
    id_anggota,
    nama,
    alamat
)

# Simpan anggota
perpus.tambah_anggota(anggota)

print("\nData anggota berhasil ditambahkan!")


# ==========================================
# TAMPILKAN DAFTAR BUKU
# ==========================================

perpus.tampil_buku()


# ==========================================
# INPUT TRANSAKSI PEMINJAMAN
# ==========================================

print("\n===== TRANSAKSI PEMINJAMAN =====")

kode_transaksi = input("Kode Transaksi : ")

# Membuat transaksi
transaksi = TransaksiPeminjaman(
    kode_transaksi,
    anggota
)

# Validasi jumlah jenis buku
while True:

    try:

        jumlah_pinjam = int(
            input("Berapa jenis buku yang dipinjam? ")
        )

        break

    except ValueError:

        print("Harap isi dengan angka!")


# Loop peminjaman
for i in range(jumlah_pinjam):

    print(f"\nPeminjaman ke-{i+1}")

    kode_cari = input(
        "Masukkan kode buku yang dipinjam : "
    )

    # Validasi jumlah pinjam
    while True:

        try:

            jumlah = int(
                input("Jumlah pinjam : ")
            )

            break

        except ValueError:

            print(
                "Harap isi jumlah pinjam "
                "dengan angka!"
            )

    # Cari buku berdasarkan kode
    for buku in perpus.daftar_buku:

        if buku.kode_buku == kode_cari:

            detail = DetailPeminjaman(
                buku,
                jumlah
            )

            transaksi.tambah_detail(detail)

# Simpan transaksi
perpus.tambah_transaksi(transaksi)

print("\nTransaksi berhasil disimpan!")


# ==========================================
# OUTPUT SISTEM
# ==========================================

print("\n===================================")
print("        OUTPUT SISTEM")
print("===================================")

# Tampilkan buku
perpus.tampil_buku()

# Tampilkan anggota
perpus.tampil_anggota()

# Tampilkan transaksi
for transaksi in perpus.daftar_transaksi:

    transaksi.tampil_transaksi()
```
